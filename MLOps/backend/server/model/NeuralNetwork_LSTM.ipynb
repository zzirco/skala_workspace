```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU, Bidirectional

import math
from sklearn.metrics import mean_squared_error
```

```python
# Some functions to help out with
def plot_predictions(test,predicted):
    plt.plot(test, color='red',label='Real IBM Stock Price')
    plt.plot(predicted, color='blue',label='Predicted IBM Stock Price')
    plt.title('IBM Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('IBM Stock Price')
    plt.legend()
    plt.show()

def return_rmse(test,predicted):
    rmse = math.sqrt(mean_squared_error(test, predicted))
    print("The root mean squared error is {}.".format(rmse))
```

```python
# Connectiing Google Drive
# from google.colab import drive
# drive.mount('/content/drive')

# Loading the data
# dataset = pd.read_csv(r'C:\phoenix\Svc\hckrs\code\model_serving\server\uploaded_files\IBM_2006-01-01_to_2018-01-01.csv', index_col='Date', parse_dates=['Date'])
dataset = pd.read_csv(r'C:\skala_workspace\MLOps\model_serving_win\server\uploaded_files\IBM_2006-01-01_to_2018-01-01.csv', index_col='Date', parse_dates=['Date'])
dataset.head()
```

```python
# Checking for missing values
# 2016ë…„ë„ì´ì „ ë°ì´í„°ì—ì„œ 2ë²ˆì§¸ ì»¬ëŸ¼ ê°’ì„ ê°€ì ¸ì™€ì„œ ì €ìž¥í•©ë‹ˆë‹¤.
training_set = dataset[:'2016'].iloc[:,1:2].values
# 2017ë…„ë„ì´í›„ ë°ì´í„°ì—ì„œ 2ë²ˆì§¸ ì»¬ëŸ¼ ê°’ì„ ê°€ì ¸ì™€ì„œ ì €ìž¥í•©ë‹ˆë‹¤.
test_set = dataset['2017':].iloc[:,1:2].values

# We have chosen 'High' attribute for prices. Let's see what it looks like
dataset["High"][:'2016'].plot(figsize=(16,4),legend=True)
dataset["High"]['2017':].plot(figsize=(16,4),legend=True)
plt.legend(['Training set (Before 2017)','Test set (2017 and beyond)'])
plt.title('IBM stock price')
plt.show()
```

```python
# Scaling the training set
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)

# Since LSTMs store long term memory state, we create a data structure with 60 timesteps and 1 output
# So for each element of training set, we have 60 previous training set elements
X_train = []
y_train = []
# LSTM ì‹œê³„ì—´ ë¶„ì„ì„ ìœ„í•´, Timestepsë¥¼ 60ìœ¼ë¡œ í•  ê²ƒì´ë¯€ë¡œ 60í–‰ ì´í›„ ë°ì´í„°ë¶€í„° ì‚¬ìš© ê°€ëŠ¥
# ë°ì´í„° ì…‹ì„ ë³´ë©´, 2016ë…„ ë°ì´í„°ê°€ ì „ì²´ 2769ê°œ ì¡´ìž¬í•¨
for i in range(60,2769): 
    X_train.append(training_set_scaled[i-60:i,0])
    y_train.append(training_set_scaled[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshaping X_train for efficient modelling
# Timestepsì™€ MinMaxScaler ë“±ì„ ì ìš©í•œ ëŒ€ìƒ ë°ì´í„°ë¥¼, LTSM Layer ìž…ë ¥ì— ìš©ì´í•˜ê²Œ ìž¬ë°°ì¹˜
# shape[0]ëŠ” ì „ì²´ í–‰ì˜ ê°¯ìˆ˜, shape[1]ì€ ì „ì²´ ì—´ì˜ ê°¯ìˆ˜ì¸ë°.. ìƒê¸° np ë°°ì—´ì— ë„£ì€ í˜•íƒœë¡œ ì—¬ê¸°ì„œëŠ” ìž˜ë¼ë‚¸ Timesteps ê°¯ìˆ˜
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
# X_train = np.reshape(X_train, (2709,60,1))ì„ ëœ»í•˜ë©°, 2016ë…„ì´ì „ 2709ê°œì˜ í–‰ì„ 60ê°œì”© ëŠì–´ë‚¸ ê°’ 1í–‰ìœ¼ë¡œ 
# LSTM (ì‹œê³„ì—´ì„± ë°ì´í„° ë³€ê²½ ì´ë ¥ ì˜ˆì¸¡ì— íŠ¹í™”ëœ RNNëª¨ë¸) input_shapeì— ë§žê²Œ 3ì°¨ì› ë°°ì—´ë¡œ ìž¬êµ¬ì„±

# The LSTM architecture
from keras.models import Sequential
from keras.layers import Input, LSTM, Dropout, Dense

regressor = Sequential()

# ìž…ë ¥ ì •ì˜ (X_train: (samples, timesteps, features) ê°€ì •)
regressor.add(Input(shape=(X_train.shape[1], 1)))

# First LSTM layer with Dropout regularisation
# 50ê°œì˜ ë…¸ë“œë¥¼ ì‚¬ìš©í•˜ì—¬ RNN, LSTM ë°©ì‹ìœ¼ë¡œ ìž…ë ¥ì¸µ, ì€ë‹‰ì¸µ ìƒì„±
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
# ì „ì²´ ê°€ëŠ¥í•œ ë°ì´í„° ë²”ìœ„ëŠ” 2709í–‰ì— ëŒ€í•´ì„œ 60ê°œ ë‹¨ìœ„ë¡œ ìžë™ ê³„ì‚°ë˜ë¯€ë¡œ ìƒëžµ
# 60ê°œì”© ëŠì–´ë‚´ì„œ, 1ê°œì˜ Featureì— ëŒ€í•´ì„œ ì˜ˆì¸¡

# Second LSTM layer
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

# Third LSTM layer
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

# Fourth LSTM layer
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))

# The output layer
regressor.add(Dense(units=1))
# ìµœì¢… ê²°ê³¼ê°€ 60ê°œì”© ì‹œê³„ì—´ ì´ë ¥ ë°ì´í„°ë¥¼ ëŠì–´ë‚¸ ë‹¨ìœ„ì— ëŒ€í•´ í•™ìŠµí•œ ê²°ê³¼, ê°’ ì˜ˆì¸¡ì´ë¯€ë¡œ ë§ˆì§€ë§‰ ë…¸ë“œëŠ” 1ê°œë¡œ ì²˜ë¦¬
# ë‹¨, MNIST ì´ë¯¸ì§€ ë°ì´í„°ì™€ ê°™ì´, 0~9ê¹Œì§€ í‘œí˜„ì„ ìœ„í•´ ìµœì¢… ë…¸ë“œë¥¼ ì‚¬ìš©í•˜ëŠ” ê²½ìš° 10ê°œ ë…¸ë“œ í•„ìš”

# Compiling the RNN
regressor.compile(optimizer='rmsprop', loss='mean_squared_error')

# Fitting to the training set
regressor.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Now to get the test set ready in a similar way as the training set.
# The following has been done so forst 60 entires of test set have 60 previous values which is impossible to get unless we take the whole
# 'High' attribute data for processing
dataset_total = pd.concat((dataset["High"][:'2016'],dataset["High"]['2017':]),axis=0)
inputs = dataset_total[len(dataset_total)-len(test_set) - 60:].values
inputs = inputs.reshape(-1,1)
inputs  = sc.transform(inputs)

# Preparing X_test and predicting the prices
X_test = []
for i in range(60,311):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)

# Visualizing the results for LSTM
plot_predictions(test_set,predicted_stock_price)
```

```python
# Evaluating our model
return_rmse(test_set,predicted_stock_price)
```
