```python
# .env íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
from dotenv import load_dotenv
load_dotenv()

import os
import time
from langchain_groq import ChatGroq
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# í™˜ê²½ë³€ìˆ˜ë¡œ Groq API Key ì„¤ì •
os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY", "ì—¬ê¸°ì—_ë³¸ì¸_KEYë¥¼_ì…ë ¥í•˜ì„¸ìš”")

# Groq Chat ëª¨ë¸ ì„¤ì • (í…ìŠ¤íŠ¸ ì „ìš© ì‘ì—…ì—ëŠ” llama3-8b-8192 ë“± ì‚¬ìš© ê¶Œì¥)
chat = ChatGroq(
    temperature=0.1,
    model="llama3-8b-8192",  # ë˜ëŠ” "llama3-70b-8192"
)

# ì›¹ í˜ì´ì§€ ë¡œë“œ
url = "https://wow.groq.com/why-groq/"
loader = WebBaseLoader(url)
docs = loader.load()

# ì˜ˆì™¸ ì²˜ë¦¬: ë¬¸ì„œê°€ ì—†ì„ ê²½ìš° ì¤‘ë‹¨
if not docs:
    raise ValueError("â— ë¬¸ì„œ ë¡œë“œ ì‹¤íŒ¨: í˜ì´ì§€ì—ì„œ ë‚´ìš©ì„ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

# ì§ˆë¬¸ ë° í”„ë¡¬í”„íŠ¸ êµ¬ì„±
question = "Groq ì—”ì§„ì˜ LLM ì¶”ë¡  ì†ë„ê°€ ë¹ ë¥¸ ì´ìœ ëŠ” ë¬´ì—‡ì¸ê°€ìš”? í•œêµ­ì–´ë¡œ ë‹µë³€í•˜ì„¸ìš”."

system = "Answer the question from given contexts. Answer in Korean."
human = """
Context: {context}

---

Question: {question}
"""

# ì²´ì¸ êµ¬ì„±
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])
chain = prompt | chat | StrOutputParser()

# ì‹¤í–‰ ë° ì¶œë ¥
start = time.time()
result = chain.invoke({
    "context": docs[0].page_content,
    "question": question
})
print("ğŸ”¹ ë‹µë³€:\n", result)

end = time.time()
print(f"\nâ±ï¸ ì²˜ë¦¬ ì‹œê°„: {end - start:.2f}ì´ˆ")
```

```python
# Stuff
example_URL='https://arxiv.org/abs/1706.03762'

loader = WebBaseLoader(example_URL)
docs = loader.load()


gemma2 = ChatGroq(
    temperature=0,
    model="gemma2-9b-it",
)

summarize_prompt = ChatPromptTemplate.from_messages([
    ('system', '''Summarize the following paper in Korean.
Emphasize the uniqueness and contribution of the paper.
Answer should be in 10 sentences.
Please Answer in Korean.
'''),
    ('user', '''{text}''')])

print(len(docs[0].page_content))
summarize_prompt.format_messages(text=docs[0].page_content)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
import urllib.request

paper_URL = "https://arxiv.org/pdf/1706.03762"


# ì™¸ë¶€ ë§í¬ì—ì„œ PDF íŒŒì¼ì„ ë‹¤ìš´ë¡œë“œí•˜ëŠ” ì½”ë“œ
urllib.request.urlretrieve(
    paper_URL,
    filename="paper-attention.pdf"
)
path = './paper-attention.pdf'
```

```python
from langchain_core.documents import Document


loader = PyMuPDFLoader("paper-attention.pdf")
# í˜ì´ì§€ë³„ë¡œ ì €ì¥
pages = loader.load()

# ì½”í¼ìŠ¤ì— ëª¨ë‘ ê²°í•©
corpus = Document(page_content='')
for page in pages:
    corpus.page_content += page.page_content
len(corpus.page_content)
```

ê¸´ Contextë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•´, ì „ì²´ ì½”í¼ìŠ¤ë¥¼ ì‘ì€ ë‹¨ìœ„ë¡œ ìª¼ê°œ ë³´ê² ìŠµë‹ˆë‹¤.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000) #chunk_sizeëŠ” í•œ chunkdp í¬í•¨ë  ìµœëŒ€ ê¸€ììˆ˜ì…ë‹ˆë‹¤.
document_list = text_splitter.split_documents([corpus])
len(document_list)
```

## Map-Reduce

```python
from tqdm import tqdm

# Map ê³¼ì • : ê° ë¬¸ì„œì— ëŒ€í•´ ìš”ì•½ì„ ìƒì„±í•©ë‹ˆë‹¤. -> Long Context ì¸ ê²½ìš°, ê· ì¼í•˜ê²Œ ë¶„í• í•˜ì—¬ ìš”ì•½í•˜ê³  ë‹¤ì‹œ í•©ì¹˜ëŠ” ë°©ë²•ìœ¼ë¡œ ì„±ëŠ¥ ê°œì„ 

map_prompt = ChatPromptTemplate.from_messages([
    ('system', '''ì£¼ì–´ì§„ ë…¼ë¬¸ì˜ ë‚´ìš©ì„ ì½ê³  í•œêµ­ì–´ë¡œ ìš”ì•½í•˜ì„¸ìš”.
ìš”ì•½ì€ 1ê°œì˜ ë¬¸ë‹¨ê³¼ ë¬¸ë‹¨ë³„ 6ê°œì˜ ë¬¸ì¥ìœ¼ë¡œ ì‘ì„±í•˜ì„¸ìš”.
ë‹µë³€ì€ í•œêµ­ì–´ë¡œ ì‘ì„±í•˜ì„¸ìš”.
ê°™ì€ ë‚´ìš©ì„ ë°˜ë³µí•˜ì§€ ë§ˆì„¸ìš”.
'''),
    ('user', '''{text}''')])

map_chain  = map_prompt | gemma2 | StrOutputParser()

raw_summaries = []
for i in tqdm(range(len(document_list))):
    response = map_chain.invoke(document_list[i].page_content)
    raw_summaries.append(response)
    print('')
    print('#',i)
    print(response)
    print('===========================')
```

```python
raw_summaries
```

```python
# Reduce ê³¼ì • : ê° ë¬¸ì„œì˜ ìš”ì•½ì„ í•˜ë‚˜ë¡œ í•©ì¹©ë‹ˆë‹¤.
reduce_prompt = ChatPromptTemplate.from_messages([

    ('system', '''
Generate a summary of the following text that includes the following elements:

* A title that accurately reflects the content of the text.
* An introduction paragraph that provides an overview of the topic.
* Bullet points that list the key points of the text.
* A conclusion paragraph that summarizes the main points of the text.

Answer in Korean.
'''),
    ('user', '''{text}
---
ìš”ì•½(In Korean):
''')])

reduce_chain = reduce_prompt | gemma2 | StrOutputParser()

summary = reduce_chain.invoke('\n---\n'.join(raw_summaries))

print(summary)
```

## Refine

Refineì€ ì²­í¬ë¥¼ ìˆœì„œëŒ€ë¡œ ì°¸ê³ í•˜ë©°, ë§¤ ì‹œì  ìš”ì•½ë¬¸ì„ ì‘ì„±í•©ë‹ˆë‹¤.   
ìš”ì•½ë¬¸ê³¼ ìƒˆë¡œìš´ ì²­í¬ë¥¼ ë¹„êµí•˜ì—¬, ìš”ì•½ë¬¸ì„ ì—…ë°ì´íŠ¸í•©ë‹ˆë‹¤.

```python
# ì²­í¬ ìˆœì„œëŒ€ë¡œ ê° ì²­í¬ë“¤ì„ ìš”ì•½í•˜ê³ , ì•ë¶€ë¶„ ì²­í¬ ìš”ì•½ê³¼ í˜„ì¬ ì²­í¬ë¥¼ ì—°ê²°í•˜ì—¬ ìš”ì•½ì„ ì—…ë°ì´íŠ¸í•˜ë©° ì§„í–‰í•œë‹¤.
# ì‹œê°„ì€ ì˜¤ë˜ ê±¸ë¦´ ìˆ˜ ìˆìœ¼ë‚˜, ì²­í¬ ì—…ë°ì´íŠ¸ê°€ ì¦ì€ ê²½ìš°ì— ì‚¬ìš©
first_summarize_prompt = ChatPromptTemplate.from_messages([
    ('system', '''ë‹¹ì‹ ì€ ì¸ê³µì§€ëŠ¥ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.
ì£¼ì–´ì§„ ë…¼ë¬¸ì˜ ë‚´ìš©ì„ ì½ê³  í•œêµ­ì–´ë¡œ ìš”ì•½í•˜ì„¸ìš”.
ìš”ì•½ì€ 1ê°œì˜ ë¬¸ë‹¨ê³¼ ë¬¸ë‹¨ë³„ 5ê°œì˜ ë¬¸ì¥ìœ¼ë¡œ ì‘ì„±í•˜ì„¸ìš”.
ë‹µë³€ì€ í•œêµ­ì–´ë¡œ ì‘ì„±í•˜ì„¸ìš”.'''),
    ('user', '''{text}''')])



X = first_summarize_prompt.format_messages(text=document_list[0])

intermediate_summary = gemma2.invoke(X).content
print(intermediate_summary)
```

```python
# Refine Prompt

refine_prompt = ChatPromptTemplate.from_messages([
    ('system', '''ë‹¹ì‹ ì€ ì¸ê³µì§€ëŠ¥ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.
ë…¼ë¬¸ì˜ í˜„ì¬ ì‹œì ê¹Œì§€ì˜ í•œêµ­ì–´ ìš”ì•½ì´ ì£¼ì–´ì§‘ë‹ˆë‹¤.
ì´ë¥¼ ì½ê³ , ìƒˆë¡­ê²Œ ì£¼ì–´ì§€ëŠ” ë‚´ìš©ê³¼ ë¹„êµí•˜ì—¬ í•œêµ­ì–´ ìš”ì•½ì„ ë³´ì™„í•˜ê±°ë‚˜ ìˆ˜ì •í•˜ì„¸ìš”.
ì „ì²´ ìš”ì•½ì€ 10ë¬¸ì¥ ì´ë‚´ë¡œ ì‘ì„±í•˜ì„¸ìš”.
'''),
    ('user', '''í˜„ì¬ ì‹œì ê¹Œì§€ì˜ ìš”ì•½: {previous_summary}
---
ìƒˆë¡œìš´ ë‚´ìš©: {new_text}''')])


refine_chain = refine_prompt | gemma2 | StrOutputParser()

for i in tqdm(range(1, len(document_list))):
    intermediate_summary = refine_chain.invoke(
        {'previous_summary':intermediate_summary,
         'new_text':document_list[i].page_content})
    print('')
    print(intermediate_summary)
    print('=======================')
# ê¸¸ì´ë¥¼ ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ ì˜¤ë˜ ê±¸ë¦´ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```
